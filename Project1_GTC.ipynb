##  Data Quality Report  Findings

After performing Exploratory Data Analysis (EDA), several data quality issues were identified:

1. **Missing Values**
   - `company`: Over 112593 missing values (>50% of the column). This makes it uninformative.
   - `agent`: A large number of missing values, but the column is still useful. Missing values were replaced with 0.
   - `country`: Contains some missing values, imputed with the mode (most frequent country).
   - `children`: Few missing values, imputed with the median.

2. **Outliers**
   - Extreme values were detected in key numerical columns (`adr`, `lead_time`).
   - Other columns also had unrealistic values, such as:
     - `adults`, `children`, `babies`  bookings with very large numbers of guests.
     - `booking_changes`, `days_in_waiting_list`, `required_car_parking_spaces`  values far beyond realistic ranges.
   - Outliers in `adr` and `lead_time` were capped using the IQR method to prevent skewing the analysis.

3. **Duplicates**
   - Exact duplicate rows were found and removed.

4. **Data Types**
   - Date columns (e.g., `reservation_status_date`) were not in datetime format and had to be converted.
   - Binary variables like `is_repeated_guest` contained unexpected values and were corrected to 0/1 only.

5. **Data Leakage**
   - `reservation_status` and `reservation_status_date` provide future information (after booking). Keeping them would cause data leakage in predictive modeling, so both were dropped.

6. **Data Duplicate**
   -We found 32,298 exact duplicate rows out of ~58,000 (55% of the dataset).  
Since duplicates do not add any new information and may bias the model, we removed them.  
This reduced the dataset size to ~26,000 unique rows, which ensures a cleaner and more reliable dataset..
   


---

###  Main Data Quality Issues
- High proportion of missing data in `company`  dropped.  
- Missing values in `agent`, `country`, and `children`  imputed.  
- Significant outliers in `adr` and `lead_time`  capped using IQR.  
- Presence of duplicates  removed.  
- Incorrect data types  fixed.  
- Data leakage from `reservation_status` variables  dropped.  

```python
df["company"] = df["company"].fillna("None")
df["agent"] = df["agent"].fillna(0)
df["country"] = df["country"].fillna(df["country"].mode()[0])
df["children"] = df["children"].fillna(df["children"].median())
```

```python
def cap_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return np.where(series > upper_bound, upper_bound,
           np.where(series < lower_bound, lower_bound, series))

df["adr"] = cap_outliers(df["adr"])
df["lead_time"] = cap_outliers(df["lead_time"])

df["reservation_status_date"] = pd.to_datetime(df["reservation_status_date"])

```

```python
print(f"Original shape: {df.shape}")
df2 = df.drop_duplicates()
print(f"After dropping duplicates: {df2.shape}")
```

```python
df2["total_guests"] = df["adults"] + df["children"] + df["babies"]
df2["total_nights"] = df["stays_in_weekend_nights"] + df["stays_in_week_nights"]
df2["is_family"] = df["children"] + df["babies"] > 0
```

```python
from sklearn.model_selection import train_test_split

df2 = df2.drop(["reservation_status", "reservation_status_date"], axis=1)
df2 = pd.get_dummies(df2, columns=["meal", "market_segment", "deposit_type", "customer_type"], drop_first=True)

country_freq = df2["country"].value_counts().to_dict()
df["country"] = df2["country"].map(lambda x: country_freq.get(x, 0))


train_df, test_df = train_test_split(df2, test_size=0.2, random_state=42)

print(" Final dataset prepared!")
print("Train shape:", train_df.shape)
print("Test shape:", test_df.shape)
```
